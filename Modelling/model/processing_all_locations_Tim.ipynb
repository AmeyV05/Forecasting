```python
import os 
import sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import processing
import importlib
```

```python
all_loc_file='../processed_data/All Locations/'
file1=all_loc_file+'total_2007_tm_2011.csv'
file2=all_loc_file+'total_2012_tm_2016.csv'
file3=all_loc_file+'total_2017_tm_2022.csv'
df1=pd.read_csv(file1)
df2=pd.read_csv(file2)
df3=pd.read_csv(file3)
df1.head()
```

```python
#getting one artikel TS
# artikel_code='AHEK'
artikel_code_vec=['J16-90','PB6090','TB80120G','A01-80-(30)','C02-80','AHEK']
for artikel_code in artikel_code_vec:    
    artikel_ind1=np.where(df1.Code.values==artikel_code)
    artikel_ind2=np.where(df2.Code.values==artikel_code)
    artikel_ind3=np.where(df3.Code.values==artikel_code)
    df1a=df1.loc[artikel_ind1].reset_index(drop=True)
    df2a=df2.loc[artikel_ind2].reset_index(drop=True)
    df3a=df3.loc[artikel_ind3].reset_index(drop=True)
    dfa=pd.concat([df1a,df2a,df3a])
    Datevec=pd.to_datetime(dfa.Datum).dt.date
    # print(Datevec)
    Net_Art=dfa['Value']
    # print(Net_Art)
    month_dict={'Date':Datevec,'Net_daily_art':Net_Art}
    # artikel_code_vec=[artikel_code]
    # artikel_code=artikel_code_vec[0]
    start_year=2007
    df_month=pd.DataFrame.from_dict(month_dict)
    file=os.path.join(all_loc_file,artikel_code+'_day'+str(start_year)+'.csv')
    df_month.to_csv(file)
```
